```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['Embarked'] = le.fit_transform(df['Embarked'])
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
df.isnull().sum()
```

```python
df.drop(['PassengerId','Name','Cabin','Fare1','Ticket','Age','Parch'],axis = 1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['Sex'] = labelencoder.fit_transform(df['Sex'])
df['Embarked'] = labelencoder.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['Survived'],axis = 1)
y = df.Survived
X_train,X_valid,y_train,y_valid = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RandomizedSearchCV

#C = np.logspace(0, 4, num=10)
#penalty = ['l1', 'l2']
#solver = ['liblinear', 'saga']
#max_iter = [100, 1000,2500, 5000]
#hyperparameters = dict(C=C, penalty=penalty, solver=solver)
#model = LogisticRegression()

#randomizedsearch = RandomizedSearchCV(model, hyperparameters)
#best_model_random = randomizedsearch.fit(X_train, y_train)
#print(best_model_random.best_estimator_)


from catboost import CatBoostClassifier
"""
parameters = {'depth'         : [4,5,6,7,8,9, 10],
                 'learning_rate' : [0.01,0.02,0.03,0.04],
                  'iterations'    : [100,200]
                 }

model = CatBoostClassifier()

randomizedsearch = RandomizedSearchCV(model, parameters)
best_model_random = randomizedsearch.fit(X_train, y_train)
best_model_random.best_params_
"""
```

```python
#from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

model1 = CatBoostClassifier(iterations=200,
                         depth=7,
                       learning_rate=0.01,
                     loss_function='Logloss',
                  verbose=True, random_state=0)
#model2  =RandomForestClassifier()
#model3 = LogisticRegression()
model1.fit(X_train,y_train)
#model2.fit(X_train,y_train)
#model3.fit(X_train,y_train)
```

```python
y_pred1 = model1.predict(X_valid)
#y_pred2 = model2.predict(X_valid)
#y_pred3 = model3.predict(X_valid)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_valid, y_pred1))
#print("Accuracy:",metrics.accuracy_score(y_valid, y_pred2))
#print("Accuracy:",metrics.accuracy_score(y_valid, y_pred3))
```

```python
metrics.classification_report(y_valid,y_pred1)
#metrics.classification_report(y_valid,y_pred2)
#metrics.classification_report(y_valid,y_pred3)
metrics.confusion_matrix(y_valid,y_pred1)
```
